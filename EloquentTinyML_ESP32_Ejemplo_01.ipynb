```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
# Cargar dataset
df = pd.read_csv('sensor_dataset.csv')
# Obtener caracteristicas de entrenamiento y etiquetas
x_train = df.iloc[:, [0]]
y_train = df.iloc[:, [1]]
```

```python
# Crear modelo 

model = keras.Sequential([
    keras.layers.Flatten(input_shape=[1]),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',             #also try adam
              metrics=['accuracy'])
```

```python
# Entrenar modelo

history = model.fit(x_train, y_train,epochs=1000)
```

```python
# Convertir modelo al formato tensorflow lite
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Guardar modelo en el disco
open("moisture_sensor_model.tflite", "wb").write(tflite_model)
```
